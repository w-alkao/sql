### Intro to views

```python
%%sql # create a view with columns vendor name, state, and phone

CREATE VIEW Vendors_min AS
  SELECT vendor_name, vendor_state, vendor_phone
  FROM Vendors;
```

```python
%%sql # select all data from the view we created

SELECT * FROM Vendors_min;
```

```python
%%sql # select all data where vendor state is CA

SELECT * FROM Vendors_min
WHERE vendor_state = 'CA'
ORDER BY vendor_name
```

```python
%%sql # 

SELECT * FROM Vendors_min
WHERE vendor_name = 'Register of Copyrights';
```

```python
%%sql # update a views, set a vendor phone for 'register of copyrights'

UPDATE Vendors_min
SET vendor_phone = '(800) 555-3941'
WHERE vendor_name = 'Register of Copyrights'
```

```python
%%sql #

SELECT * FROM Vendors_min
WHERE vendor_name = 'Register of Copyrights';
```

```python
%%sql # drop a view we created

DROP VIEW Vendors_min;
```

```python
%%sql 

SELECT * FROM Vendors_min
```
