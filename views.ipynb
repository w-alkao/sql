### Work with view

```python
%%sql # create a view of vendors that have invoices

CREATE VIEW Vendor_phone_list AS
  SELECT vendor_name, vendor_contact_last_name,
    vendor_contact_first_name, vendor_phone
  FROM Vendors
  WHERE vendor_id IN (
    SELECT DISTINCT vendor_id
    FROM Invoices
  );

SELECT * FROM Vendor_phone_list;
```

```python
%%sql # create a view that uses a join, it select all vendors that are also in Invoices tables

CREATE OR REPLACE VIEW Vendor_invoices As
  SELECT vendor_name, invoice_number, invoice_date, invoice_total
  FROM Vendors v
    JOIN Invoices i ON v.vendor_id = i.vendor_id;

SELECT * FROM Vendor_invoices;
```

```python
%%sql # create a view that uses a LIMIT clause, with the top 5 highest invoices totals

CREATE OR REPLACE VIEW Top5_invoice_totals AS
  SELECT vendor_id, invoice_total
  FROM Invoices
  ORDER BY invoice_total DESC
  LIMIT 5;

SELECT * FROM Top5_invoice_totals;
```

```python
%%sql # create a view that names all of its columns

CREATE OR REPLACE VIEW Invoice_outstanding (i_number, i_date, i_total, balance_due) AS
  SELECT invoice_number, invoice_date, invoice_total,
    invoice_total - payment_total - credit_total AS balance_due
  FROM Invoices
  HAVING balance_due > 0;

SELECT * FROM Invoice_outstanding;
    
```

```python
%%sql # create a view that names just the calculated column in its select clause

CREATE OR REPLACE VIEW Invoice_outstanding AS
  SELECT invoice_number, invoice_date, invoice_total AS i_total,
    invoice_total - payment_total - credit_total AS balance_due
  FROM Invoices
  HAVING balance_due > 0;

SELECT * FROM Invoice_outstanding;
```

```python
%%sql # create a view that summarizes invoices by vendor

CREATE OR REPLACE VIEW Invoice_summary AS 
  SELECT vendor_name, COUNT(*) AS i_count, SUM(invoice_total) AS i_total_sum
  FROM Vendors v
    JOIN Invoices i ON v.vendor_id = i.vendor_id
  GROUP BY vendor_name;

SELECT * FROM Invoice_summary;
```

#### Create an updatable view

##### Requierements for creating updatable views
* The select list can't include a DISTINCT clause;
* The select list can't include aggregate functions;
* The SELECT statement can't include a GROUP BY or HAVING clause;
* The veiw can't include the UNION operator.

```python
%%sql # create an updatable view

CREATE OR REPLACE VIEW Balance_due_view AS
  SELECT vendor_name, invoice_number,
    invoice_total, payment_total, credit_total,
    invoice_total - payment_total - credit_total AS balance_due
  FROM Vendors v
    JOIN Invoices i ON v.vendor_id = i.vendor_id
  WHERE invoice_total - payment_total - credit_total > 0;

SELECT * FROM Balance_due_view;
```

```python
%%sql # use the view to update data

UPDATE Balance_due_view
SET credit_total = 300
WHERE invoice_number = "9982771";

SELECT * FROM Balance_due_view;
```

```python
%%sql # attempt to use the view to update a calculated column

UPDATE Balance_due_view
SET balance_due = 0
WHERE invoice_number = "9982771";
```

```python
%%sql # updatable view that has a WITH CHECK OPTION clause

CREATE OR REPLACE VIEW Vendor_payment AS
  SELECT vendor_name, invoice_number, invoice_date, payment_date,
    invoice_total, credit_total, payment_total
  FROM Vendors v
    JOIN Invoices i ON v.vendor_id = i.vendor_id
  WHERE invoice_total - payment_total - credit_total >= 0
WITH CHECK OPTION;

SELECT * FROM Vendor_payment;
```

```python
%%sql # display a row from the view

SELECT * FROM Vendor_payment
WHERE invoice_number = "P-0608";
```

```python
%%sql # update the view

UPDATE Vendor_payment
SET payment_total = 400.00, payment_date = "2018-08-01"
WHERE invoice_number = "P-0608";

SELECT * FROM Vendor_payment
WHERE invoice_number = "P-0608";
```

```python
%%sql # attempt to update the view

UPDATE Vendor_payment
SET payment_total = 30000.00, payment_date = "2018-08-01"
WHERE invoice_number = "P-0608";
```

```python
%%sql #1

CREATE OR REPLACE VIEW open_items AS
SELECT vendor_name, invoice_number, invoice_total,
  invoice_total - payment_total - credit_total AS balance_due
FROM Vendors v
  JOIN Invoices i ON v.vendor_id = i.vendor_id
HAVING balance_due > 0
ORDER BY vendor_name;

SELECT * FROM open_items;
```

```python
%%sql #2

SELECT * FROM open_items
WHERE balance_due >= 1000;
```

```python
%%sql #3

CREATE OR REPLACE VIEW open_items_summary AS
SELECT vendor_name, COUNT(*) AS open_item_count,
  SUM(invoice_total - credit_total - payment_total) AS open_item_total
FROM Vendors v
  JOIN Invoices i ON v.vendor_id = i.vendor_id
WHERE invoice_total - credit_total - payment_total > 0
GROUP BY vendor_name
ORDER BY open_item_total DESC;

SELECT * FROM open_items_summary;
```

```python
%%sql #

SELECT * FROM open_items_summary
LIMIT 5;
```
