### create test tables (copy of a table)

```python
%%sql #create a complete copy of the invoices tables

CREATE TABLE Invoices_copy AS
SELECT * FROM Invoices
```

```python
%sql DESCRIBE Invoices_copy
```

```python
%sql SELECT * FROM Invoices_copy
```

```python
%%sql # Create a partial copy of the Invoices table a table with no balance due

CREATE TABLE old_invoices AS
SELECT * FROM Invoices
WHERE invoice_total - payment_total - credit_total = 0
```

```python
%sql SELECT * FROM old_invoices;
```

```python
%%sql #Create a table with summary rows from the Invoices table

CREATE TABLE Vendor_balances AS
SELECT vendor_id, SUM(invoice_total) AS sum_of_invoices
FROM Invoices
WHERE (invoice_total - payment_total - credit_total) <> 0
GROUP BY vendor_id;
```

```python
%sql SELECT * FROM Vendor_balances;
```

### Updata existing rows

```python
%sql SELECT * FROM Invoices
```

```python
%%sql #Update two columns for a single row

UPDATE Invoices
SET payment_date = "2018-09-21", payment_total = 19351.18
WHERE invoice_number = '97/522'
```

```python
%%sql #update one column for multiple rows

UPDATE Invoices
SET terms_id = 1
WHERE vendor_id = 95
```

```python
%%sql #update one column for one row

UPDATE Invoices
SET credit_total = credit_total + 100
WHERE invoice_number = '97/522'
```

```python
%%sql #update the terms for all invoices for vendors in three states

UPDATE Invoices
SET terms_id = 1
WHERE vendor_id IN
    (SELECT vendor_id
     FROM Vendors
     WHERE vendor_state IN ('CA', 'AZ', 'NV'))
```

```python
%%sql #update all invoices for a vendor

UPDATE Invoices
SET terms_id = 1
WHERE vendor_id = (
  SELECT vendor_id
  FROM Vendors
  WHERE vendor_name = 'Pacific Bell')
```

### Delete existing rows

```python
%%sql #use a subquery in a delete statement

DELETE FROM Invoice_line_items
WHERE invoice_id IN (
  SELECT invoice_id
  FROM Invoices
  WHERE vendor_id = 115);
```
