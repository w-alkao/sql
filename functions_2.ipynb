```python
%sql SELECT * FROM Invoices;
```

```python
%%sql # select statement that uses a simple CASE function

SELECT invoice_number, terms_id,
  CASE terms_id
    WHEN 1 THEN 'Net due 10 days'
    WHEN 2 THEN 'Net due 20 days'
    WHEN 3 THEN 'Net due 30 days'
    WHEN 4 THEN 'Net due 40 days'
    WHEN 5 THEN 'Net due 50 days'
  END AS terms
FROM Invoices;
```

```python
%%sql # select statement that uses a searched CASE function

SELECT invoice_number, invoice_total, invoice_date, invoice_due_date,
  CASE
    WHEN DATEDIFF('2018-09-13	', invoice_due_date) > 30
      THEN 'Over 30 days past due'
    WHEN DATEDIFF('2018-09-13	', invoice_due_date) > 0
      THEN '1 to 30 days past due'
    ELSE 'Current'
  END AS invoice_status
FROM Invoices
WHERE invoice_total - payment_total - credit_total > 0;
```

```python
%%sql # IF function: return vendor name and yes if vendor city is fresno and no else

SELECT vendor_name, vendor_city,
  IF(vendor_city = 'Fresno', 'Yes', 'No') AS is_city_fresno
FROM Vendors;

```

```python
%%sql # IFNULL function: return payment date and subtitues null values with 'No payment'

SELECT payment_date,
  IFNULL(payment_date, 'No payment') AS new_date
FROM Invoices;
```

```python
%%sql # COALESCE function: return payment date and substitue null value with 'No payment'

SELECT payment_date,
  COALESCE(payment_date, 'No payment') AS new_date
FROM Invoices;
```

#### Regular expression functions

```python
%%sql # r0: is 23 in the string ?
      # r1: is the string starting with ab ?
      # r2: is the string ending with 23 ?
      # r3: is the single value 2 in the string ?
      # r4: is is at least one element of list [ad] in the string ?

SELECT REGEXP_LIKE('abc123', '23') AS r0,
  REGEXP_LIKE('abc123', '^ab') AS r1,
  REGEXP_LIKE('abc123', '23$') AS r2,
  REGEXP_LIKE('abc123', '.2') AS r3,
  REGEXP_LIKE('abc123', '[ad]') as r4;
```

```python
%%sql # r0: is 24 in the string ?
      # r1: is the string starting with bc ?
      # r2: is the string ending with 12 ?
      # r3: is the single value 9 in the string ?
      # r4: is is at least one element of list [fd] in the string ?

SELECT REGEXP_LIKE('abc123', '24') AS r0,
  REGEXP_LIKE('abc123', '^bc') AS r1,
  REGEXP_LIKE('abc123', '12$') AS r2,
  REGEXP_LIKE('abc123', '.9') AS r3,
  REGEXP_LIKE('abc123', '[fd]') as r4;
```

```python
%%sql #

SELECT REGEXP_LIKE('abc123', '[a-f]') AS r0,
  REGEXP_LIKE('abc123', 'b|d') AS r1,
  REGEXP_LIKE('abc123', 'd*') AS r2,
  REGEXP_LIKE('abc123', '(bcd)*') as r4;
```

```python
%%sql #

SELECT REGEXP_LIKE('abc123', '[d-f]') AS r0,
  REGEXP_LIKE('abc123', 'z|d') AS r1,
  REGEXP_LIKE('abc123', 'd*') AS r2,
  REGEXP_LIKE('abc123', '(bcd)*') as r4;
```

```python
%%sql # Return the index of the first character of the substring in the string expression
      # that matches the pattern

SELECT REGEXP_INSTR('abc123', '[2-5]') AS r0,
  REGEXP_INSTR('abc123', 'b|d') AS r1,
  REGEXP_INSTR('abc123', 'f*') AS r2,
  REGEXP_INSTR('abc123', '(bcd)*') as r4;
```

```python
%%sql # return the first substring of the string expression that matches the pattern

SELECT REGEXP_SUBSTR('abc123', '[A-Z][1-9]*$') AS r0;
```

```python
%%sql # return the string expression with all occurences of the pattern replaced with the replace string

SELECT REGEXP_REPLACE('abc123', '1|3', '*') AS ro;
```

```python
%%sql # return vendor city and the position of the first spaces in vendor city

SELECT DISTINCT vendor_city,
  REGEXP_INSTR(vendor_city, ' ') AS space_index
FROM Vendors
HAVING space_index > 0
```

```python
%%sql # return vendor city where the name of city start with SAN or LOS

SELECT vendor_city,
  REGEXP_SUBSTR(vendor_city, '^SAN|LOS') AS city_match
FROM Vendors
HAVING city_match IS NOT NULL;
```

```python
%%sql # return vendor name and address with the word street in their name
      # and replace this word with St for new address

SELECT vendor_name, vendor_address1,
  REGEXP_REPLACE(vendor_address1, 'STREET', 'St') AS new_address1
FROM Vendors
WHERE REGEXP_LIKE(vendor_address1, 'STREET');
```

#### Ranking functions

```python
%%sql # sort the result set by vendor name and number each row in the result set

SELECT ROW_NUMBER() OVER(ORDER BY vendor_name) AS 'row_number', vendor_name
FROM Vendors;
```

```python
%%sql # sort result by vendor name within each state and number each row within these stat

SELECT ROW_NUMBER() OVER(PARTITION BY vendor_state
    ORDER BY vendor_name) AS 'row_number', vendor_name, vendor_state
FROM Vendors;
```

```python
%%sql #

SELECT RANK() OVER(ORDER BY invoice_total) AS 'rank',
  DENSE_RANK() OVER(ORDER BY invoice_total) AS 'dense_rank',
  invoice_total, invoice_number
FROM Invoices;
```

```python
%%sql #

SELECT terms_description, terms_id,
  NTILE(2) OVER(ORDER BY terms_id) AS tile2,
  NTILE(3) OVER(ORDER BY terms_id) AS tile3,
  NTILE(4) OVER(ORDER BY terms_id) AS tile4
FROM Terms;
```

```python
%%sql # how to use FIRST, NTH AND LAST VALUES

SELECT sales_year, CONCAT(rep_first_name, ' ', rep_last_name) AS rep_name,
  sales_total,
  FIRST_VALUE(CONCAT(rep_first_name, ' ', rep_last_name)) OVER
    (PARTITION BY sales_year ORDER BY sales_total DESC) AS highest_sales,
  NTH_VALUE(CONCAT(rep_first_name, ' ', rep_last_name), 3) OVER
    (PARTITION BY sales_year ORDER BY sales_total DESC RANGE BETWEEN
    UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS third_highest_sales,
  LAST_VALUE(CONCAT(rep_first_name, ' ', rep_last_name)) OVER
    (PARTITION BY sales_year ORDER BY sales_total DESC RANGE BETWEEN
    UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS lowest_sales
FROM Sales_totals st
  JOIN Sales_reps sr ON st.rep_id = sr.rep_id;
```

```python
%%sql # Lag FUNCTION

SELECT rep_id, sales_year, sales_total AS current_sales,
  LAG(sales_total, 1, 0) OVER (PARTITION BY rep_id ORDER BY sales_year) AS last_sales,
  sales_total - LAG(sales_total, 1, 0) OVER (PARTITION BY rep_id ORDER BY sales_year) AS 'change'
FROM Sales_totals;
```

```python
%%sql # how to use PERCENT_RANK and CUME_DIST functions

SELECT sales_year, rep_id, sales_total,
  PERCENT_RANK() OVER (PARTITION BY sales_year ORDER BY sales_total) AS pct_rank,
  CUME_DIST() OVER (PARTITION BY sales_year ORDER BY sales_total) AS 'cume_dist'
FROM Sales_totals;
```
