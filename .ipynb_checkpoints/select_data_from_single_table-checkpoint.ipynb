```python
%%sql

#Select all male or parrot but not both

SELECT * FROM Animal WHERE sex="M"
  XOR species="parrot";
```

```python
%%sql

#Select all animal either born after 2019 nor cat male or female,
#if female cat they must born before 2017

SELECT * FROM Animal
WHERE birth_date > "2019-01-01" ||
  (species="cat" && (sex="M" || (sex="F" && birth_date < "2017-06-01")))
```

## Basic select statement

```python
%%sql
# Retrieve all data from the invoices tables

SELECT * FROM Invoices;
```

```python
%%sql
# Retrieve three columns from each each row, sorted in descending sequence by invoice_total

SELECT invoice_number, invoice_date, invoice_total
FROM Invoices
ORDER BY invoice_total DESC
```

```python
%%sql
# Retrieve two columns and a calculated value for a specific invoice

SELECT invoice_id, invoice_total, credit_total+payment_total AS total_credits
FROM Invoices
WHERE invoice_id = 17
```

```python
%%sql
# Retrieve all invoice between given dates

SELECT invoice_number, invoice_date, invoice_total
FROM Invoices
WHERE invoice_date BETWEEN "2018-06-01" AND "2018-06-30"
ORDER BY invoice_date
```

```python
%%sql
# Statement that returns an empty result set

SELECT invoice_number, invoice_date, invoice_total
FROM Invoices
WHERE invoice_total > 50000
```

## Using alias

```python
%%sql
# Statement that renames the columns in the result set

SELECT invoice_number AS "Invoice Number",
       invoice_date AS Date, 
       invoice_total AS Total
FROM Invoices
```

```python
%%sql
# Statement that doesn't name a calculated column

SELECT invoice_number, invoice_date, invoice_total,
       invoice_total - payment_total - credit_total
FROM Invoices
```

## Arithmetic expressions

```python
%%sql # Statement that calculates the balance due

SELECT invoice_total, payment_total, credit_total,
       invoice_total - payment_total - credit_total AS balance_due
FROM Invoices
```

```python
%%sql # Arithmetich operations

SELECT invoice_id,
       invoice_id + 7 * 3 AS multiply_first,
       (invoice_id + 7) * 3 AS add_first
FROM Invoices
ORDER BY invoice_id
```

```python
%%sql # Div and modulo operators

SELECT invoice_id,
       invoice_id / 3 AS decimal_quotient,
       invoice_id DIV 3 AS integer_quotient,
       invoice_id % 3 AS remainder
FROM Invoices
ORDER BY invoice_id
```

## Concat function to join strings

```python
%%sql # Concatenate string data

SELECT vendor_city, vendor_state, CONCAT(vendor_city, vendor_state)
FROM Vendors
```

```python
%%sql # Format string data using litaral values

SELECT vendor_name,
       CONCAT(vendor_city, ", ", vendor_state, " ", vendor_zip_code) AS address
FROM Vendors
```

```python
%%sql # Using concatenation again

SELECT CONCAT(vendor_name, "'s Address: ") AS Vendors,
       CONCAT(vendor_city, ", ", vendor_state, " ", vendor_zip_code) AS Address
FROM Vendors
```

### use functions with strings, dates, and numbers

```python
%%sql # left function, to split some character from left

SELECT vendor_contact_first_name, vendor_contact_last_name,
       CONCAT(LEFT(vendor_contact_first_name, 1),
              LEFT(vendor_contact_last_name, 1)) AS initials
FROM Vendors
```

```python
%%sql # Right function, to split some character from right

SELECT vendor_contact_first_name, vendor_contact_last_name,
       CONCAT(RIGHT(vendor_contact_first_name, 2), " ",
              RIGHT(vendor_contact_last_name, 2)) AS initials
FROM Vendors
```

```python
%%sql # Specify a format of date with date_format

SELECT invoice_date,
       DATE_FORMAT(invoice_date, "%m/%d/%y") AS "MM/DD/YY",
       DATE_FORMAT(invoice_date, "%e-%b-%Y") AS "DD-Month-YYYY"
FROM Invoices
ORDER BY invoice_date
```

```python
%%sql # Round function

SELECT invoice_date, invoice_total,
       ROUND(invoice_total) AS nearest_dollar,
       ROUND(invoice_total, 1) AS nearest_dime
FROM Invoices
ORDER BY invoice_date
```

### Eliminate duplicate rows

```python
%%sql # select all vendors

SELECT vendor_city, vendor_state
FROM Vendors
ORDER BY vendor_city
```

```python
%%sql # Eliminate all duplicate row 

SELECT DISTINCT vendor_city, vendor_state
FROM Vendors
ORDER BY vendor_city
```

### Comparison operators

```python
%%sql # Vendors located in lowa

SELECT * FROM Vendors
WHERE vendor_state = "IA"
```

```python
%%sql # invoice with a balance due

SELECT * FROM Invoices
WHERE invoice_total - payment_total - credit_total > 0
```

```python
%%sql # Vendor name from A to L

SELECT vendor_name FROM Vendors
WHERE vendor_name < "M"
```

```python
%%sql # invoice with credits that don't equal zero

SELECT * FROM Invoices
WHERE credit_total != 0
```

#### AND, OR, NOT, IN, BETWEEN operator

```python
%%sql # vendor in NJ state and East Brunswick city

SELECT vendor_id, vendor_name, vendor_state, vendor_city
FROM Vendors
WHERE vendor_state = "NJ" AND vendor_city = "East Brunswick"
```

```python
%%sql # vendor in NJ state or Fresno city

SELECT vendor_id, vendor_name, vendor_state, vendor_city
FROM Vendors
WHERE vendor_state = "NJ" OR vendor_city = "Fresno"
```

```python
%%sql # Vendor whose in not in CA state

SELECT vendor_id, vendor_name, vendor_state
FROM Vendors
WHERE NOT vendor_state = "CA"
```

```python
%%sql # in operator

SELECT * FROM Invoices
WHERE terms_id IN (1, 2, 4)
```

```python
%%sql # in operator

SELECT vendor_id, vendor_name, vendor_state
FROM Vendors
WHERE vendor_state NOT IN ("CA", "NV", "OR")
```

```python
%%sql #vendor with invoices

SELECT vendor_id, vendor_name, vendor_state, vendor_city
FROM Vendors
WHERE vendor_id IN (
  SELECT vendor_id
  FROM Invoices
)
```

```python
%%sql # between op

SELECT * FROM Invoices
WHERE payment_total BETWEEN credit_total AND credit_total + 500
```

```python
%%sql # between op

SELECT * FROM Invoices
WHERE payment_total NOT BETWEEN credit_total AND credit_total + 500
```

### the use of like and regexp

```python
%%sql # city where name state with San

SELECT vendor_city
FROM Vendors
WHERE vendor_city LIKE "SAN%"
```

```python
%%sql # name which start with "compu" follow by one character and "er"

SELECT vendor_name
FROM Vendors
WHERE vendor_name LIKE "COMPU_ER%"
```

```python
%%sql #city with "sa" in their name

SELECT vendor_city
FROM Vendors
WHERE vendor_city REGEXP "SA"
```

```python
%%sql # city wich name start with "sa"

SELECT vendor_city FROM Vendors
WHERE vendor_city REGEXP "^SA"
```

```python
%%sql # city wich name end with "sa"

SELECT vendor_city FROM Vendors
WHERE vendor_city REGEXP "NA$"
```

```python
%%sql # city with "rs" or "sn" in their name

SELECT vendor_city FROM Vendors
WHERE vendor_city REGEXP "RS|SN"
```

```python
%%sql # city with "n" followed by  "c" or "v" in their name

SELECT vendor_city FROM Vendors
WHERE vendor_city REGEXP "N[CV]"
```

```python
%%sql # city with "N" followed with A to J in their name

SELECT vendor_city FROM Vendors
WHERE vendor_city REGEXP "N[A-J]"
```

```python
%%sql 

SELECT vendor_contact_last_name FROM Vendors
WHERE vendor_contact_last_name REGEXP "DAMI[EO]N"
```

```python
%%sql

SELECT invoice_due_date AS "Due Date",
       invoice_total AS "Invoice Total",
       invoice_total * .1 AS "10%",
       invoice_total * (1 + .1) AS "Plus 10%"
FROM Invoices
WHERE invoice_total BETWEEN 500 AND 1000
ORDER BY invoice_due_date DESC
```

```python
%%sql

SELECT invoice_number,
       invoice_total,
       payment_total + credit_total AS payment_credit_total,
       invoice_total - payment_total - credit_total AS balance_due
FROM Invoices
WHERE invoice_total - payment_total - credit_total > 50
ORDER BY balance_due DESC
LIMIT 5
```
