## Inner join

```python
%%sql #how to join two tables

SELECT invoice_number, vendor_name
FROM Vendors INNER JOIN Invoices
  ON Vendors.vendor_id = Invoices.vendor_id
ORDER BY invoice_number
```

```python
%%sql # join tables using alias

SELECT invoice_number, vendor_name, invoice_due_date,
       invoice_total - payment_total - credit_total AS balance_due
FROM Vendors V JOIN Invoices I
  ON V.vendor_id = I.vendor_id
WHERE invoice_total - payment_total - credit_total > 0
ORDER BY invoice_due_date DESC
```

```python
%%sql # using alias with on table

SELECT invoice_number, line_item_amount, line_item_description
FROM Invoices JOIN Invoice_line_items LI
  ON Invoices.invoice_id = LI.invoice_id
WHERE account_number =540
ORDER BY invoice_date
```

```python
%%sql # join table in another database

SELECT vendor_name, customer_last_name, customer_first_name,
       vendor_state AS state, vendor_city AS city
FROM Vendors v JOIN om.Customers c
  ON v.vendor_zip_code = c.customer_zip
ORDER BY state, city
```

```python
%%sql # compound join conditions

SELECT customer_first_name, customer_last_name
FROM om.Customers c JOIN ex.Employees e
  ON c.customer_first_name = e.first_name
  AND c.customer_last_name = e.last_name
```

```python
%%sql # self-join, select vendors from cities in common with other vendors

SELECT DISTINCT v1.vendor_id, v1.vendor_name, v1.vendor_city, v1.vendor_state
FROM Vendors v1 JOIN Vendors v2
  ON v1.vendor_city = v2.vendor_city
  AND v1.vendor_state = v2.vendor_state
  AND v1.vendor_name <> v2.vendor_name
ORDER BY v1.vendor_id
```

```python
%%sql # join more than two tables

SELECT vendor_name, invoice_number, invoice_date, line_item_amount,
       account_description
FROM Vendors v
  JOIN Invoices i ON v.vendor_id = i.vendor_id
  JOIN Invoice_line_items li ON i.invoice_id = li.invoice_id
  JOIN General_ledger_accounts gl ON li.account_number = gl.account_number
WHERE invoice_total - payment_total - credit_total > 0
ORDER BY vendor_name, line_item_amount DESC
```

```python
%%sql # implicit inner join

SELECT invoice_number, vendor_name
FROM Vendors v, Invoices i
WHERE v.vendor_id = i.vendor_id
ORDER BY invoice_number
```

```python
%%sql # join four table with implicit join

SELECT vendor_name, invoice_number, invoice_date, line_item_amount,
       account_description
FROM Vendors v, Invoices i, Invoice_line_items li, General_ledger_accounts gl
WHERE v.vendor_id = i.vendor_id
  AND i.invoice_id = li.invoice_id
  AND li.account_number = gl.account_number
  AND invoice_total - payment_total - credit_total > 0
ORDER BY vendor_name, line_item_amount DESC
```

## other skills for joins

```python
%sql USE ap
```

```python
%%sql # join tables with the USING

SELECT invoice_number, vendor_name
FROM Vendors JOIN Invoices USING (vendor_id)
```

```python
%%sql # join three table with USING

USE ex;

SELECT department_name, last_name, project_number
FROM Departments
  JOIN Employees USING (department_number)
  LEFT JOIN Projects USING (employee_id)
```

```python
%%sql # join table with natural keyword

USE ap;

SELECT invoice_number, vendor_name 
FROM Vendors NATURAL JOIN Invoices
```

```python
%%sql #use natural join to join three table

SELECT department_name AS dept_name, last_name, project_number
FROM ex.Departments 
  NATURAL JOIN ex.Employees
  LEFT JOIN ex.Projects USING (employee_id)
ORDER BY department_name
```

```python
%%sql # cross join

SELECT d.department_number, department_name, employee_id, last_name
FROM ex.Departments d CROSS JOIN ex.Employees
ORDER BY d.department_number
```

```python
%%sql #implicit cross join

SELECT d.department_number, department_name, employee_id, last_name
FROM ex.Departments d, ex.Employees
ORDER BY d.department_number
```

## Work with unions

```python
%%sql

SELECT 'Active' AS source,
        invoice_number, invoice_date, invoice_total
FROM ex.Active_invoices
WHERE invoice_date >= '2018-06-01'
```

```python
%%sql 

SELECT 'Paid' AS source, invoice_number, invoice_date, invoice_total
FROM ex.Paid_invoices
WHERE invoice_date >= '2018-06-01'
```

```python
%%sql #Combine results set from two different tables

SELECT 'Active' AS source, invoice_number, invoice_date, invoice_total
FROM ex.Active_invoices
WHERE invoice_date >= '2018-06-01'

UNION

SELECT 'Paid' AS source, invoice_number, invoice_date, invoice_total
FROM ex.Paid_invoices
WHERE invoice_date >= '2018-06-01'
```

```python
%%sql # combines result set from the same tables

SELECT 'Active' AS source,
       invoice_number, invoice_date, invoice_total
FROM Invoices
WHERE invoice_total - payment_total - credit_total > 0

UNION

SELECT 'Paid' AS source,
       invoice_number, invoice_date, invoice_total
FROM Invoices
WHERE invoice_total - payment_total - credit_total <= 0        
```

```python
%%sql # combine result set from the same two tables

SELECT invoice_number, vendor_name, '33% Payment' AS payment_type,
       invoice_total AS total, invoice_total * 0.333 AS payment
FROM Invoices i JOIN Vendors v ON i.vendor_id = v.vendor_id
WHERE invoice_total > 10000

UNION

SELECT invoice_number, vendor_name, '50% Payment' AS payment_type,
       invoice_total AS total, invoice_total * 0.5 AS payment
FROM Invoices i JOIN Vendors v ON i.vendor_id = v.vendor_id
WHERE invoice_total BETWEEN 500 AND 10000

UNION

SELECT invoice_number, vendor_name, 'Full amount' AS payment_type,
       invoice_total AS total, invoice_total AS payment
FROM Invoices i JOIN Vendors v ON i.vendor_id = v.vendor_id
WHERE invoice_total < 500

ORDER BY payment_type, vendor_name, invoice_number
```

```python
%%sql # similates a full outer join with union

SELECT department_name AS dept_name, d.department_number AS d_dept_no,
       e.department_number AS e_dept_no, last_name
FROM ex.Departments d LEFT JOIN ex.Employees e
  ON d.department_number = e.department_number

UNION

SELECT department_name AS dept_name, d.department_number AS d_dept_no,
       e.department_number AS e_dept_no, last_name
FROM ex.Departments d RIGHT JOIN ex.Employees e
  ON d.department_number = e.department_number
```
