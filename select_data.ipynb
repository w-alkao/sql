```python
%%sql

-- Animal born before 2018

SELECT * FROM Animal WHERE birth_date < "2018-01-01";
```

```python
# select all animal except cat

%sql SELECT * FROM Animal WHERE species <> "cat";
```

```python
%%sql 

#Select all female cats

SELECT * FROM Animal
WHERE species = "cat" && sex = "F";
```

```python
%%sql

# Select all turtle or parrot

SELECT * FROM Animal WHERE species = "turtle" || species = "parrot"
```

```python
%%sql

#Select * female except dog

SELECT * FROM Animal WHERE sex="F" && !(species="dog");
```

```python
%%sql

#Select all male or parrot but not both

SELECT * FROM Animal WHERE sex="M"
  XOR species="parrot";
```

```python
%%sql

#Select all animal either born after 2019 nor cat male or female,
#if female cat they must born before 2017

SELECT * FROM Animal
WHERE birth_date > "2019-01-01" ||
  (species="cat" && (sex="M" || (sex="F" && birth_date < "2017-06-01")))
```

```python
%%sql

#Select all animals without name

SELECT * FROM Animal WHERE name IS NULL;
```

```python
%%sql

#Select all animals where comments is set

SELECT * FROM Animal WHERE comments IS NOT NULL
```
